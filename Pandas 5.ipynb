```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first_name": ['Syed', 'Syed', 'Umer', 'Ateeb', 'Mr', 'Mr'],
    "middle_name": ['Dayim', 'Ayub', 'Lateef', 'Showkat', 'C', 'C'],
    "last_name": ['Shah', 'Shah', 'Naiko', 'Bhanghi', 'Lang', '++'],
    "contact": ['111', '666', '555', '444', '333', '222'],
}

mydata = pd.DataFrame(people)
mydata
```

```python
mydata.rename(columns={
    'first_name': 'fn',
    'middle_name': 'mn',
    'last_name': 'ln',
    'contact': 'C',
}, inplace=True)

mydata
```

```python
mydata.loc[4, ['ln']] = 'Shah'
mydata
```

```python
fltr = (mydata['fn'] == 'Mr') & (mydata['ln'] == 'Shah')
mydata.loc[fltr, 'C'] = 0

mydata
```

# Applying changes to multiple rows at once

### Method 1: apply
##### it works on a series

```python
def update_C(C):
    C = 0
    return C

mydata['C'] = mydata['C'].apply(update_C)

mydata
```

### applymap()
##### it works on a dataFrame, not on series

```python
mydata.loc[0:,['fn', 'mn', 'ln']].applymap(str.upper)
```
